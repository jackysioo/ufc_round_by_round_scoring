```python
import pandas as pd
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
import requests
from bs4 import BeautifulSoup as bs
```

```python
# load the projectpro webpage content 

r = requests.get('https://www.ufc.com/news/official-judges-scorecards-ufc-286-edwards-vs-usman-3') 



# convert to beautiful soup 

soup = bs(r.content)



# printing our web page 

print(soup.prettify()) 
```

```python
images = soup.select('div img') 

images_url = images[0]['src'] 
```

```python
image_urls = []
for image in images:
    image_urls.append(image['src'])
```

```python
image_urls = [image['src'] for image in images]
```

```python
image_urls
```

```python
img_url = image_urls[0]
```

```python
img_bytes = requests.get(img_url).content 
```

```python
type(img_bytes)
```

```python
import io
from PIL import Image

imageStream = io.BytesIO(img_bytes)
imageFile = Image.open(imageStream)
```

```python
type(imageFile)
```

```python
import numpy as np
img_array = np.array(imageFile)
```

```python
img_array.shape
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img_array)
```

```python
import math
num_images = len(image_urls)
rows = 5
cols = math.ceil(num_images / rows)

fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(20, 20))
axes = axes.reshape(-1)

for i, img_url in enumerate(image_urls):
    try:
        img_bytes = requests.get(img_url).content 
        imageStream = io.BytesIO(img_bytes)
        imageFile = Image.open(imageStream)
        img_array = np.array(imageFile)

        axes[i].imshow(img_array)
        axes[i].set_title(f"img_array.shape = {img_array.shape}")
    except:
        print("Failed")
```
